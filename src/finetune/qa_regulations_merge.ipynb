```python
from pathlib import Path

input_files = [
    Path("./data/regulations_ab_strict_cleaned.jsonl"),
    Path("./data/regulations_c.jsonl"),
    Path("./data/regulations_df_final.jsonl"),
]

output_file = Path("./data/regulations_merged_for_finetuning.jsonl")

with output_file.open("w", encoding="utf-8") as outfile:
    for file_path in input_files:
        with file_path.open("r", encoding="utf-8") as infile:
            for line in infile:
                line = line.strip()
                if line:  # 빈 줄 제외
                    outfile.write(line + "\n")

print(f"합치기 완료: {output_file}")
```
